```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import shutil
```

```python
entries = []
indices = []
col_names = ['customer',
 'revenue',
 'customer-revenue',
 'unweighted-network-features',
 'weighted-network-features',
 'customer-revenue-unweighted-network-features',
 'customer-revenue-weighted-network-features',
 ]
d = '.\\yoshi\\output\\last6m-median'
for f in os.listdir(d):
    if 'fimp' not in f:
        df = pd.read_csv(os.path.join(d, f))
        entries.append(df.iloc[5,1:].to_list())
        fname = f.split('.')[0]
        ind = "-".join(fname.split('_')[2:]).replace('demographic', 'customer').replace('first6m-rev-mod', 'revenue')
        indices.append(ind)

res = pd.DataFrame(entries, index=indices, columns=df.columns[1:])

ax = res.loc[col_names[::-1]].plot.barh(rot=0, width=0.7)
ax.set_xlim([0.6, 0.7])
fig_name = d.split('\\')[-1]
ax.set_title(fig_name)

plt.savefig(f'{fig_name}.png', bbox_inches='tight')
```
